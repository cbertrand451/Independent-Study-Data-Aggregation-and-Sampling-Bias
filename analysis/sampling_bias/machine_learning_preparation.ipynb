## Check other grouping structure

Species is the main grouping variable in this lesson, but it is not the only structure in the data. `sex`, `plot_type`, and seasonality can also affect what a sample represents.

```python
sex_distribution = (
    sample_df['sex']
    .value_counts(dropna=False, normalize=True)
    .rename_axis('sex')
    .reset_index(name='full_dataset_percent')
)

plot_type_distribution = (
    sample_df['plot_type']
    .value_counts(dropna=False, normalize=True)
    .rename_axis('plot_type')
    .reset_index(name='full_dataset_percent')
)

month_distribution = (
    sample_df['month']
    .value_counts(normalize=True)
    .sort_index()
    .rename_axis('month')
    .reset_index(name='full_dataset_percent')
)

sex_distribution, plot_type_distribution, month_distribution
```

These distributions help students see that sampling is not only about row counts. A sample can accidentally overrepresent a sex, plot type, season, or year if the sampling strategy does not match the research question.
